## Most common words

Firstly, let's solve the original challenge. What are the 100 most common words in Pride and Prejudice?

Note: Before I start I've saved the Pride and Prejudice text file at "p_and_p.txt" in my current working directory.

```python
# I'm going to use the following modules for my analysis
# They are standard Python libraries. You shouldn't need to install anything extra
import string
import collections


with open("p_and_p.txt") as f: # open the file for reading
    text = f.read() # read file into a single super long string
    
    # I'd like to remove all punctuation from the text and make everything lower case
    text = text.translate(str.maketrans('', '', string .punctuation)).lower()
    
    # I now want to split the string into a massive list of words
    words = text.split()
    
    # Now lets count how many of each word there is in this list. the collections module is brill at this.
    counter = collections.Counter(words)
    
    # counter.most_comon() will generate a list of tuples containing ("word", frequency) pairs
    # I can slice these to get the top 100.
    most_common = counter.most_common()[:100]
```

```python
# I'm going to use the amazing pandas module
# to get my data into a form that I can easily
# save to a file or plot
import pandas as pd

# create a dataframe with column headers "word" and "frequency"
df = pd.DataFrame(most_common, columns=['word', 'frequency'])

# you can save this really easily to an excel spreadsheet using
df.to_excel("p_and_p_most_common_words.xlsx")

# how about a bar chart of the 20 most common words?
```

## Let's make a bar chart

```python
# how about a bar chart of the 20 most common words?
# we can use pandas along with matplotlib to do this

# don't forget to import the matplotlib module
import matplotlib.pyplot as plt

# create a new data frame with only the first 20 rows from df
df2 = df.head(20)
df2.plot(kind='bar', x='word')

# cardinal sin if you don't label the axis!
plt.ylabel("Frequency")

plt.show() # need this to make the figure appear
```

## Let's make a word cloud

```python
# the wordcloud module is not a native Python module.
# you will need to install it using pip or your package manager
# worth the effor though, right?

from wordcloud import WordCloud

with open("p_and_p.txt") as f:
    text = f.read() # same as before
    wordcloud = WordCloud(background_color="white").generate(text)

    # Display the generated image:
    # the matplotlib way:
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    
    plt.show()
```
